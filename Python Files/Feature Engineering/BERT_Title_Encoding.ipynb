```python
from sentence_transformers import SentenceTransformer
import pandas as pd
from IPython.display import clear_output
import numpy as np
import glob
```

```python
EMBED_OUTPUT = 768
```

```python
model = SentenceTransformer("bert-base-nli-mean-tokens", device="cuda")
```

```python
def generateLabels():
  labels = []
  for i in range(EMBED_OUTPUT):
    labels.append(f"embed_{i+1}")
  return labels
```

```python
def generateTitleEmbeddings(df):
  titles = list(df["videoTitle"])
  labels = generateLabels()
  embeddings = model.encode(titles)
  df[labels] = embeddings
  return df
```

```python
fileList = glob.glob("./data_2022-05-08/**/**/X*.csv")
```

```python
for idx, fileName in enumerate(fileList):
  clear_output(wait=True)
  print(f"{idx+1}/{len(fileList)}")
  df = pd.read_csv(fileName)
  df = df.loc[:, ~df.columns.str.contains("^Unnamed")]
  df = generateTitleEmbeddings(df)
  df.to_csv(fileName.replace("test", "test_bert").replace("train","train_bert"))
```
