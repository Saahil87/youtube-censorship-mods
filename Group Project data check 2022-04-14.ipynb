# Unremoved videos

```python
# unremoved videos for channel: UCBi2mrWuNuyYy4gbM6fU18Q
df_unremoved = pd.read_csv('new.csv', encoding='utf8', low_memory=False, lineterminator='\n', delimiter='\t')
```

```python
print(df_unremoved.shape)
print(df_unremoved.columns)
```

```python
# 19,794 unique videos from ABC news
df_unremoved['videoId'].value_counts()
```

```python
# all but one video is ABC, but one is from "25" for some reason
df_unremoved['channelId'].value_counts()

# we remove the bad row
df_unremoved = df_unremoved[df_unremoved['channelId'] != '25']
print(df_unremoved.shape)
```

```python
# change date column to first ten digits
df_unremoved['uploadDate'] = df_unremoved['publishedAt'].apply(lambda x: x[:10])

# don't need time
del df_unremoved['publishedAt']
```

```python
# datetime fix
df_unremoved['uploadDate'] = pd.to_datetime(df_unremoved['uploadDate'], format='%Y-%m-%d')
```

```python
# how many are for 2021?
df_unremoved[(df_unremoved['uploadDate'] > '2021-01-01') & (df_unremoved['uploadDate'] < '2022-01-01')].shape

# 6712 rows of original 19,793
```

```python
# add signifier for removed
df_unremoved['removed'] = False
```

# Joining data

```python
# joining unremoved and removed videos
all_vids = pd.concat([df_removed, df_unremoved])
print(all_vids.shape)

# four columns were exactly the same names and thus merged
```

```python
# how many unique videos?
all_vids['videoId'].value_counts()

# 572,442
```

```python
# joining channel info and all vids (unremoved and removed)
df = pd.merge(all_vids, channel_df, on="channelId")
```

```python
# how large is merged df?
print(df.shape)

# how many unique videos?
df['videoId'].value_counts()

# we see a few videos where the count is over 1, because the video is in both the unremoved 
# (i.e. all videos) and removed videos
```
