# Fix encoding of political channel tags

```python
# make a boolean column for each tag
s = df['tags_y']
mlb = MultiLabelBinarizer()
labels_df = pd.DataFrame(mlb.fit_transform(s),columns=mlb.classes_, index=df.index)
labels_df.head()
```

```python
# join the labels boolean with the actual data
print(df.join(labels_df).shape)
df = df.join(labels_df)
```
