```python
political_categories_list = sorted(channel_df_exploded.dropna().unique().tolist())
political_categories_list
```

```python
def random_sample(political_cat, sample_n):
    print(channel_df.explode('tags')[channel_df.explode('tags')['tags']==political_cat].sample(sample_n, replace=False, random_state=2)['channelId'].to_list())
    return channel_df.explode('tags')[channel_df.explode('tags')['tags']==political_cat].sample(sample_n, replace=False, random_state=2)
```

```python
data_frame_list = []
for x in political_categories_list:
    print(x)
    data_frame_list.append(random_sample(x, 2))

sample_df = pd.concat(data_frame_list)
sample_df
```
